```python
def find_closest(x):
    try:
        c = [x['SHOT_RESULT']]
        if x['CLOSE_DEF_PERSON_ID'] != float('nan') and x["Player1ID"] != float('nan')  and  int(x["Player1ID"]) == int(x['CLOSE_DEF_PERSON_ID']):
            c.append([x["P1X"], x["P1Y"],  x["Player1Team"]])
        for idx in xrange(2,11):
            if x['CLOSE_DEF_SV_PLAYER_ID'] != float('nan') and x["Player"+str(idx)+".1"] != float('nan') and int(x["Player"+str(idx)+".1"]) == int(x['CLOSE_DEF_PERSON_ID']):
                c.append([x['CLOSE_DEF_SV_PLAYER_ID'], x["P"+str(idx)+"X"], x["P"+str(idx)+"Y"]])
                team = x["Player"+str(idx)+"Team"]
                num = idx
        for idx in xrange(1,11):
            if x["Player"+str(idx)+"Team"] == team and idx != num:
                c.append([2, x["P"+str(idx)+"X"], x["P"+str(idx)+"Y"]])
            else:
                c.append([1, x["P"+str(idx)+"X"], x["P"+str(idx)+"Y"]])
        return c
    except Exception:
        return []
```
